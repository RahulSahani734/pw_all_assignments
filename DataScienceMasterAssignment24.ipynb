### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be easily created using Seaborn along with their uses:

1. **Scatter Plot (`sns.scatterplot`)**:
   - Use: Scatter plots are used to visualize the relationship between two numerical variables. They help identify patterns, correlations, clusters, or outliers in the data.
   - Example: Studying the correlation between a student's study hours and their exam scores.

2. **Bar Plot (`sns.barplot`)**:
   - Use: Bar plots are suitable for comparing categorical data by displaying the average value of a numerical variable for each category. They allow easy comparison between categories.
   - Example: Comparing the average sales of different products across multiple regions.

3. **Line Plot (`sns.lineplot`)**:
   - Use: Line plots are used to display the trend or pattern in numerical data over a continuous interval. They are commonly used for time series data and continuous variables.
   - Example: Plotting the temperature variations throughout a day.

4. **Histogram (`sns.histplot`)**:
   - Use: Histograms display the distribution of a single numerical variable by dividing it into bins and showing the frequency of data points in each bin. They are useful for visualizing the data's underlying distribution.
   - Example: Understanding the distribution of student test scores in a class.

5. **Heatmap (`sns.heatmap`)**:
   - Use: Heatmaps are used to display the relationship between two categorical variables, represented as a 2D matrix, where each cell's color encodes the value of the relationship.
   - Example: Visualizing the correlation between different features in a dataset.

These are just a few examples of the wide range of plots that Seaborn offers. Seaborn simplifies the process of creating visually appealing and informative plots, making it a popular choice for data visualization tasks.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset from Seaborn
fmri_data = sns.load_dataset("fmri")

# Plot the line plot using Seaborn's lineplot function
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x='timepoint', y='signal', hue='event', style='region', markers=True)

# Set plot labels and title
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('Line Plot of "signal" vs. "timepoint" for different events and regions')

# Show the legend
plt.legend(title='Event')
plt.grid(True)
plt.show()
```

 ### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
 ### Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Plot two box plots side by side using Seaborn's boxplot function
plt.figure(figsize=(12, 6))

# First box plot for 'age' vs. 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age vs. Passenger Class')

# Second box plot for 'fare' vs. 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Box Plot of Fare vs. Passenger Class')

plt.tight_layout()  # To avoid overlapping of subplots
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram using Seaborn's histplot function
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)

# Set plot labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price with Cut as Hue')

# Show the legend
plt.legend(title='Cut', loc='upper right')

plt.grid(True)
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset from Seaborn
iris_data = sns.load_dataset("iris")

# Plot the pair plot using Seaborn's pairplot function
sns.pairplot(iris_data, hue='species', markers=["o", "s", "D"])

# Set plot title
plt.suptitle("Pair Plot of Iris Dataset with Species as Hue", y=1.02)

plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot the heatmap using Seaborn's heatmap function
plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', cbar_kws={'label': 'Passengers'})
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Heatmap of Flight Passengers')

plt.show()
```
