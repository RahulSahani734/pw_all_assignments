## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python, designed to be lightweight and easy to use. It provides the necessary tools and libraries to build web applications quickly and efficiently. Here are some key points about Flask:

1. Flask Framework: Flask is a Python web framework that follows the Model-View-Controller (MVC) architectural pattern. It provides a simple and minimalistic approach to web development, allowing developers to focus on the core functionality of their applications.

2. Advantages of Flask Framework:
   - Lightweight and Minimalistic: Flask has a small and simple core, which means it does not impose any specific dependencies or project structures. This gives developers the freedom to choose the tools and libraries they want to work with.
   - Easy to Learn: Flask has a straightforward and intuitive API that makes it easy for developers, especially beginners, to understand and get started with. The simplicity of the framework helps in reducing the learning curve.
   - Flexibility: Flask allows developers to customize and tailor their applications according to their specific needs. It does not enforce any rigid patterns or configurations, giving developers the freedom to design their application structure as desired.
   - Extensibility: Flask provides a wide range of extensions that can be easily integrated into applications to add additional functionality. These extensions cover various areas such as database integration, authentication, testing, and more.
   - Large Community and Ecosystem: Flask has a vibrant and active community of developers, which means there is a wealth of resources, tutorials, and third-party packages available. This makes it easier to find solutions to common problems and promotes code reusability.
   - Scalability: While Flask is considered a micro framework, it can still be used to build scalable applications. By leveraging additional tools and libraries, such as SQLAlchemy for database management and Flask Blueprints for modular application design, Flask can handle complex and larger projects effectively.

Overall, Flask's simplicity, flexibility, and extensibility make it a popular choice for developers who want a lightweight framework to build web applications quickly and efficiently.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views within a Flask application. It allows you to define different routes that users can access in your web application and define the corresponding actions or responses for each route.

When a user requests a specific URL, Flask's app routing mechanism matches that URL to the defined routes in your application and executes the associated function or view. This way, you can define how your application responds to different URLs or endpoints.

Here are a few reasons why we use app routes in Flask:

1. URL Mapping: App routes allow you to map specific URLs to functions or views in your Flask application. This helps in defining the structure and organization of your application's URLs.

2. Handling Different HTTP Methods: App routes allow you to handle different HTTP methods such as GET, POST, PUT, DELETE, etc. For example, you can define a route that responds to GET requests for displaying information and another route that responds to POST requests for submitting data.

3. Modular Application Design: By using app routes, you can break down your application into smaller, self-contained views or functions. Each route represents a specific functionality or feature, making your application more modular and easier to manage.

4. Dynamic URL Parameters: App routes support dynamic URL parameters, allowing you to create routes that can accept variable values. This enables you to create flexible and dynamic routes that can handle different data or queries.

5. Separation of Concerns: App routes help in separating the concerns of your application. You can have different routes for handling user authentication, data retrieval, data manipulation, and more. This separation promotes code organization and maintainability.

6. URL Generation: Flask provides URL generation functions that can generate URLs based on the defined app routes. This is useful when you want to create links or redirect users to specific routes within your application.

In summary, app routes in Flask are used to map URLs to specific functions or views in your application, allowing you to define the behavior and responses for each route. They provide a way to handle different HTTP methods, create modular application designs, and manage the structure and organization of your web application.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: Company Name: ABC Corporation Location: India Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation<br>
              Location: India<br>
              Contact Detail: 999-999-9999'''

if __name__=="__main__":
    app.run(host="0.0.0.0")

```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It is a helpful function that generates a URL for a specific function or endpoint based on its name.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Generating a URL for the 'hello' function
    hello_url = url_for('hello')
    return f'The URL for hello function: {hello_url}'

@app.route('/hello')
def hello():
    return 'Hello, Flask!'

if __name__ == '__main__':
    app.run()
```
