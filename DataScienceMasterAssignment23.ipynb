### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot. import numpy as np np.random.seed(3)
### x = 3 + np.random.normal(0, 2, 50)
### y = 3 + np.random.normal(0, 2, len(x))
### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y ,marker = 'o',c ='g',label='Data points')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplote')
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:
### import numpy as np
### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in matplotlib is used to create multiple plots (subplots) within a single figure. It is especially useful when you want to compare different datasets or visualize related data side by side. The subplot() function allows you to arrange plots in rows and columns, and you can specify which subplot you want to work on by providing the row, column, and index of the subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2', color='orange')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 4')
plt.legend()

plt.tight_layout()  # To avoid overlapping of subplots
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
### import numpy as np company = np.array(["Apple", "Microsoft", "Google", "AMD"]) profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of data visualization that displays data using rectangular bars of varying lengths or heights. Each bar represents a category or group, and the length or height of the bar corresponds to the value of the data for that category. Bar plots are particularly useful for visualizing and comparing discrete data or data that can be grouped into distinct categories.

Bar plots are commonly used for the following purposes:

Comparison: Bar plots allow easy comparison of values between different categories. By visualizing the lengths or heights of the bars, you can quickly see which category has higher or lower values.

Categorical Data: Bar plots are ideal for representing categorical data, such as different products, countries, or companies, where the data points fall into distinct categories.

Distribution: Bar plots can be used to display the distribution of data across different categories.

Trends and Patterns: By observing the relative heights of bars, you can identify trends and patterns in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Bar Plot - Company Profit')
plt.grid(axis='y')  # Add grid lines on the y-axis
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='green')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot - Company Profit')
plt.grid(axis='x')  # Add grid lines on the x-axis
plt.show()
```

###Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
###box1 = np.random.normal(100, 10, 200)
###box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of the data, which includes the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The box in the plot represents the interquartile range (IQR), which is the range between Q1 and Q3. The whiskers extend from the box to the minimum and maximum data points within a certain range, typically 1.5 times the IQR. Data points outside this range are shown as individual points and are considered outliers.

Box plots are useful for the following purposes:

Visualizing Distribution: Box plots provide a concise summary of the data distribution, allowing you to understand the spread and skewness of the data.

Comparison: Box plots are excellent for comparing the distribution of multiple datasets side by side.

Identification of Outliers: Box plots can easily identify outliers in the data.

Detection of Skewness: By observing the box plot, you can quickly determine if the data is symmetric or skewed.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# Create a figure and axis
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])

plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')

plt.grid(True)
plt.show()
```
